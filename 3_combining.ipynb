```python
import glob
import re

# 파일 패턴에 맞는 모든 파일 리스트를 가져옴
file_pattern = './reference/chunk/reference_chunk_*.txt'
file_list = glob.glob(file_pattern)

# 파일명에서 숫자를 추출하여 정렬하기 위한 함수
def extract_number(filename):
    s = re.findall(r'\d+', filename)
    return int(s[0]) if s else -1

# 파일 리스트를 파일명에 포함된 숫자 기준으로 정렬
file_list.sort(key=extract_number)

# 파일 리스트 순서 확인
print("처리될 파일 리스트 순서:")
for fname in file_list:
    print(fname)
    
# 최종 파일 생성
output_file = './reference/merged_references_0_to_990.txt'

# 첫 번째 파일을 포함하여 모든 파일을 순차적으로 열고 내용을 최종 파일에 추가
with open(output_file, 'w') as outfile:
    for i, fname in enumerate(file_list):
        print(f"Processing file: {fname}")  # 현재 처리 중인 파일 출력
        with open(fname, 'r') as infile:
            # 모든 내용을 읽어서 최종 파일에 쓰기
            outfile.write(infile.read())

print("모든 파일이 성공적으로 합쳐졌습니다.")

```

```python
import glob
import re

# 파일 패턴에 맞는 모든 파일 리스트를 가져옴
file_pattern = './reference/chunk2/reference_chunk_*.txt'
file_list = glob.glob(file_pattern)

# 파일명에서 가장 큰 숫자를 추출하여 정렬하기 위한 함수
def extract_number(filename):
    numbers = re.findall(r'\d+', filename)
    numbers = [int(num) for num in numbers]  # 모든 숫자를 정수로 변환
    return max(numbers) if numbers else -1  # 가장 큰 숫자 반환

# 파일 리스트를 파일명에 포함된 숫자 기준으로 정렬
file_list.sort(key=extract_number)

# 파일 리스트 순서 확인
print("처리될 파일 리스트 순서:")
for fname in file_list:
    print(fname)
```

# 최종파일 합산

최종 파일 합산 코드 예시

앞선 파일(헤더가 있는 파일)과 뒷부분 파일(헤더를 추가했던 파일)을 병합할 때, 뒷부분 파일에서 헤더를 제거하고 싶다면, 첫 번째 줄(헤더)을 건너뛰고 내용만 병합하는 방법을 사용할 수 있습니다.

이 방법을 사용하면, 최종 파일을 병합할 때는 불필요한 헤더를 제거할 수 있습니다. 

파일 병합 과정에서 메모리 에러가 발생하는 경우, 이는 파일의 내용을 한 번에 읽으려고 할 때 시스템의 사용 가능한 메모리가 부족하기 때문일 수 있습니다. 특히, 매우 큰 파일을 처리할 때 이런 상황이 발생할 수 있습니다. 이 문제를 해결하기 위해, 파일의 내용을 한 번에 전부 읽는 대신, 파일을 줄 단위로 순차적으로 읽고 쓰는 방법을 사용할 수 있습니다. 이 방식은 메모리 사용량을 크게 줄여줍니다.

```python
now = datetime.datetime.now() 
print('시작:', now)

# 앞선 파일과 뒷부분 파일 경로
first_file = './mid/third/third_result2.txt'  # 헤더가 있는 파일
second_file = './mid/third/third_result4.txt'  # 병합할 뒷부분 파일 (헤더 포함)

# 최종 병합 파일 경로
final_output_file = './mid/third/final_merged_file.txt'

# 첫 번째 파일 내용을 그대로 복사
with open(first_file, 'r') as infile, open(final_output_file, 'w') as outfile:
    for line in infile:
        outfile.write(line)

# 두 번째 파일에서 첫 줄(헤더)을 제외하고 내용 복사
with open(second_file, 'r') as infile, open(final_output_file, 'a') as outfile:
    next(infile)  # 첫 줄 건너뛰기 (헤더 제외)
    for line in infile:
        outfile.write(line)
'''
# 아래 코드는 메모리 에러 난 코드입니다. 
# 두 파일 병합, 두 번째 파일에서는 첫 줄(헤더)을 제외하고 병합
with open(final_output_file, 'w') as outfile:
    # 첫 번째 파일 내용을 그대로 복사
    with open(first_file, 'r') as infile:
        outfile.write(infile.read())
    
    # 두 번째 파일에서 첫 줄(헤더)을 제외하고 내용 복사
    with open(second_file, 'r') as infile:
        next(infile)  # 첫 줄 건너뛰기 (헤더 제외)
        outfile.write(infile.read())
'''
now = datetime.datetime.now()
print('종료:', now)
```
