문자열 'NaN'을 포함하여 데이터를 텍스트 파일로 저장하고, 이후에 Pandas로 다시 읽을 때 발생하는 이 현상은 데이터 타입 추론과 결측치 처리 방식 때문입니다. Pandas는 데이터를 읽을 때 각 열의 데이터 타입을 추론하려고 시도하며, 특정 열에 숫자와 문자열이 혼합되어 있을 경우 처리 방식에 따라 다르게 작동할 수 있습니다.

문제 상황 설명
문자열로 처리된 '0', '1', 'NaN': 데이터를 저장할 때 'within_core' 열의 값들을 문자열로 처리하여 '0', '1', 'NaN' 등으로 저장했습니다.
텍스트 파일 저장 시 공백 표시: 텍스트 파일로 저장하면서 문자열 'NaN'이 공백으로 표시되는 경우가 발생할 수 있습니다. 이는 저장 방식에 따라 다를 수 있습니다.
Pandas로 다시 읽을 때 0.0, 1.0으로 바뀜: 저장된 파일을 Pandas의 read_csv 등의 함수로 다시 읽을 때, Pandas는 숫자로 추론할 수 있는 데이터를 실수형으로 변환하려고 시도할 수 있습니다. 문자열 'NaN'이 결측치로 처리되고, 나머지 숫자로 간주되는 문자열은 실수로 변환될 수 있습니다. 이 과정에서 '0', '1' 등도 실수형 0.0, 1.0으로 변환됩니다.
해결 방법
저장할 때 결측치 처리: 결측치를 명시적으로 문자열 'NaN'으로 저장하려면, 저장 과정에서 이를 명확히 지정해야 합니다. CSV 파일로 저장할 때 na_rep='NaN' 옵션을 사용하여 결측치를 'NaN' 문자열로 명시적으로 표현할 수 있습니다.
데이터를 읽을 때 데이터 타입 명시: 데이터를 다시 읽을 때는 dtype 매개변수를 사용하여 특정 열의 데이터 타입을 명시적으로 지정할 수 있습니다. 예를 들어, 'within_core' 열을 문자열로 읽으려면 dtype={'within_core': str} 옵션을 사용합니다.
결측치 표현 방식 명시: read_csv 함수의 na_values 매개변수를 사용하여 파일에서 결측치로 간주될 값을 명시적으로 지정할 수 있습니다. 이렇게 하면 'NaN' 문자열을 결측치로 정확히 인식할 수 있습니다.

```python
# reading 
now = datetime.datetime.now() 
print('시작:', now)
third_reload =  pd.read_csv('./mid/third/final_merged_file.txt', sep='\t', dtype={'within_core': str}, na_values=['NaN'])
now = datetime.datetime.now()
print('종료:', now)
```
