### 데이터 타입 일치 체크
결측값을 드롭하지 않고 각 컬럼 내의 데이터 타입 일관성을 검사하는 것도 가능합니다. 하지만, Pandas에서 결측값(NaN)은 float 타입으로 취급되므로, 숫자형 컬럼에 결측값이 포함되어 있으면 해당 컬럼의 데이터 타입이 float로 인식될 수 있습니다. 이는 데이터 타입 일관성을 검사할 때 고려해야 할 중요한 점입니다.  
  
결측값의 존재 자체가 문제가 되는 것이 아니라, 결측값으로 인해 데이터 타입이 변환되거나 예상치 못한 데이터 타입이 포함되는 경우에 주의해야 합니다. 예를 들어, 정수형 컬럼에 결측값이 하나라도 포함되면, 해당 컬럼 전체가 부동소수점형(float)으로 변환됩니다.  
  
결측값을 드롭하지 않고 데이터 타입 일관성을 검사할 경우, 주로 컬럼의 dtype을 직접 검사합니다. 이 방법은 결측값의 존재 여부에 관계없이 컬럼의 데이터 타입을 확인할 수 있게 해줍니다.  

```python
now = datetime.datetime.now() 
print('시작:', now)



# 각 컬럼별로 데이터 타입 일관성 체크
for column in df1.columns:
    # 결측값이 아닌 유니크한 값들의 타입 체크
    unique_types = set(type(v) for v in df1[column].dropna().unique())
    
    # 일관성 결과 출력
    if len(unique_types) > 1:
        print(f"컬럼 '{column}'은(는) 여러 타입을 포함하고 있습니다: {unique_types}")
    else:
        print(f"컬럼 '{column}'은(는) 일관된 타입을 가지고 있습니다: {unique_types}")

now = datetime.datetime.now()
print('종료:', now)
```

```python
# 각 컬럼의 데이터 타입 출력
for column in df1.columns:
    print(f"컬럼 '{column}'의 데이터 타입: {df1[column].dtype}")
```
